```
import pandas as pd
import numpy as np
from scipy import stats
```

```
df = pd.read_csv("/content/health_dataset.csv")
```

#1. Hypothesis

```
contingency = pd.crosstab(df['smoking_status'], df['diabetes'])
chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("Chi-square p-value:", p)
```

# 2. Confidence Interval (Age)


```
mean = np.mean(df['age'])
std = np.std(df['age'], ddof=1)
n = len(df)

ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std/np.sqrt(n))
print("Confidence Interval (Age):", ci)
```

#4. z-test / t-test

```
male = df[df['gender']=='Male']['bmi']
female = df[df['gender']=='Female']['bmi']

t_stat, p_val = stats.ttest_ind(male, female)
print("T-test p-value:", p_val)
```

# 5. Chi-square Test (Categorical)

```
contingency = pd.crosstab(df['smoking_status'], df['diabetes'])

chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("Chi-square p-value:", p)
```

#6. ANOVA Test (Age group vs Glucose)

```
groups = [group['glucose_level'].values for name, group in df.groupby('age_group')]

f_stat, p_anova = stats.f_oneway(*groups)

print("ANOVA p-value:", p_anova)
```

#7. Covariance & Correlation

```
cov = df['age'].cov(df['bmi'])
corr = df['age'].corr(df['bmi'])

print("Covariance:", cov)
print("Correlation:", corr)
```
